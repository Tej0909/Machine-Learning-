```python
# Importing Notification Pacakage(pip install plyer)
from plyer import notification
import requests
from bs4 import BeautifulSoup
import time
```

```python
# Creating function (Only .ico is supported as app_icon in this package)
def notifyMe(title,message):
    notification.notify(
        title = title,
        message = message,
        app_icon = 'C:/Users/ravii/Desktop/covid_19_featured_area_icon_Q9J_icon.ico',
        timeout = 15
    )
```

```python
def getData(url):
    data = requests.get(url)
    return data.text
```

```python
Html_Extract = getData('https://www.mohfw.gov.in/')
```

```python
soup = BeautifulSoup(Html_Extract,'html.parser')
```

```python
print(soup.prettify())
```

```python
my_Datastr = ""
for tr in soup.find_all('tbody')[7].find_all('tr'):
    my_Datastr += tr.get_text()
my_Datastr = my_Datastr[1:]
covid_info = my_Datastr.split('\n\n')
```

```python
states = ['Andhra Pradesh','Karnataka','Telengana','Odisha']
while True:
    for item in covid_info[0:23]:
        datalist = item.split('\n')
        if datalist[1] in states:
             print(datalist)
             nTitle = 'Cases Of COVID-19'
             ntext = f'{datalist[1]} Indian Cases: {datalist[2]}\nForgien National: {datalist[3]}\nCured/Discharged: {datalist[4]}\nDeath : {datalist[5]}'
             notifyMe(nTitle,ntext)
             time.sleep(2)
    time.sleep(3600)
    
```
